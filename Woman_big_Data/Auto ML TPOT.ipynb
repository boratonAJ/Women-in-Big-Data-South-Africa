```python
# enc pipeline
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline, make_union
from tpot.builtins import StackingEstimator,OneHotEncoder

from sklearn.linear_model import RidgeCV, ElasticNetCV
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor


# Average CV score on the training set was:0.8767668328812122
pipeline_enc1 = make_pipeline(
    StackingEstimator(estimator=ElasticNetCV(l1_ratio=0.9500000000000001, tol=1e-05)),
    RandomForestRegressor(bootstrap=False, max_features=0.1, min_samples_leaf=1, min_samples_split=4, n_estimators=1000)
)

# Average CV score on the training set was:0.8630661690341384

pipeline_enc2 = RandomForestRegressor(bootstrap=False,
                                      max_features=0.25, 
                                      min_samples_leaf=2,
                                      min_samples_split=2, 
                                      n_estimators=1000)

# Average CV score on the training set was:0.8409039139854155
pipeline_enc3 = make_pipeline(
    OneHotEncoder(minimum_fraction=0.1, sparse=False, threshold=10),
    StackingEstimator(
        estimator=XGBRegressor(
            learning_rate=0.0206999,
            max_depth=3,
            min_child_weight=1,
            n_estimators=1000, 
            nthread=1,
            subsample=0.6000000000000001)),
    RidgeCV()
)

# Average CV score on the training set was:0.8784382472941576

pipeline_enc4 = make_pipeline(
    StackingEstimator(estimator=RandomForestRegressor(bootstrap=False, max_features=0.5, min_samples_leaf=2, min_samples_split=4, n_estimators=1000)),
    XGBRegressor(learning_rate=0.0206999, max_depth=3, min_child_weight=12, n_estimators=1000, nthread=1, subsample=0.8500000000000001)
)
```
